```python
weather = {
    'Day':[1,2,3,4,5,6,7],
    'Temperature':[50,51,52,48,47,49,46]
}
```

```python
import pandas as pd
df = pd.DataFrame(weather)
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('Day')
plt.ylabel('temperature')
plt.title('weather')
plt.plot( df['Day'], df['Temperature'] , color = 'red',linewidth=5 , linestyle = 'dotted')
```
